```python
import pandas as pd
df = pd.read_csv(
    "/content/BindingDB_BindingDB_Articles.tsv",
    sep="\t",
    engine="python",
    on_bad_lines="skip")
df.head()
```

```python

df = df[["BindingDB MonomerID", "IC50 (nM)"]]
df["IC50 (nM)"] = pd.to_numeric(df["IC50 (nM)"], errors="coerce")
df = df.dropna(subset=["IC50 (nM)"])

result = df.groupby("BindingDB MonomerID")["IC50 (nM)"].agg(
    Mean="mean",
    Median="median",
    Max="max",
    Min="min"
)

result.head(10).style.format("{:.2f}")
```
